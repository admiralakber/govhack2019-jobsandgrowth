# Processing

### Get 5 year wage change in % and abs at national level

```python
df_wage_national_change = df_salary_pred.transpose()
df_wage_national_change['Predicted 5 Year Wage Growth'] = np.around(df_wage_national_change[2022]-df_wage_national_change[2017], decimals=2)
df_wage_national_change['Predicted 5 Year Wage Growth (%)'] = np.around(100*(df_wage_national_change[2022]-df_wage_national_change[2017])/df_wage_national_change[2017], decimals=2)
df_wage_national_change = df_wage_national_change[['Predicted 5 Year Wage Growth', 'Predicted 5 Year Wage Growth (%)']]
df_wage_national_change.columns = ['Predicted 5-Year Wage Growth ($) (National)', 'Predicted 5-Year Wage Growth (%)']
```

```python
df_wage_national_change.head()
```

```python
df_wage_national_and_lmip = df_lmip[['Skill Level', 'Predicted New Jobs Over Next 5 Years', 'Predicted 5-year Employment Growth (%)']].copy()
df_wage_national_and_lmip.columns = ['Skill Level', 'Predicted New Jobs Over Next 5 Years (National)', 'Predicted 5-year Employment Growth (%)']
df_wage_national_and_lmip['Predicted 5-year Employment Growth (%)'] = np.around(df_wage_national_and_lmip['Predicted 5-year Employment Growth (%)'], decimals=2)
df_wage_national_and_lmip = df_wage_national_and_lmip.join(df_wage_national_change)
```

```python
df_wage_national_and_lmip.head()
```

```python
df_wage_national_and_lmip = df_wage_national_and_lmip.reset_index()
df_wage_national_and_lmip = df_wage_national_and_lmip.rename(columns={'Unit Group':'ANZSCO'})
df_wage_national_and_lmip 
```

```python
df_anzsco_complete = df_wages_state.reset_index(drop=True).set_index('ANZSCO').join(df_wage_national_and_lmip.set_index('ANZSCO') ).copy()
```

```python
df_anzsco_complete['Skill Level'] =df_anzsco_complete['Skill Level'].fillna(4)
```

```python
df_anzsco_complete['Predicted New Jobs Over Next 5 Years (State)'] = np.round((df_anzsco_complete['Predicted 5-year Employment Growth (%)']/100*df_anzsco_complete['Number of Individuals']))
df_anzsco_complete['Predicted 5 Year Wage Growth ($) (State)'] = np.round((df_anzsco_complete['Predicted 5-Year Wage Growth (%)']/100*df_anzsco_complete['Salary or wage income (mean)']))
```

```python
df_anzsco_complete.reset_index().set_index(['ANZSCO','State'])
```
