```python
import pandas as pd
import numpy as np
```

```python
df_path = './data/ATO_2016-2017_SnapshotTable7e.xlsx'
```

```python
df = pd.read_excel(df_path, sheet_name=5, skiprows=3)
```

```python
cols_new = list(df.columns)
cols_new[0] = 'ANZSCO'
df.columns  = ['ANZSCO',
               'State',
               'Number of Individuals',
               'Taxable Income (mean)',
               'Taxable Income (median)',
               'Salary or wage income (mean)',
               'Salary or wage income (median)',
               'Total Income (mean)',
               'Total Income (median)',
              ]
df['ANZSCO Name'] = df['ANZSCO'].str[5:]
df['ANZSCO'] = df['ANZSCO'].str[:4]
df['ANZSCO Minor Group'] = df['ANZSCO'].str[:3]

df = df[['ANZSCO',
         'ANZSCO Name',
         'ANZSCO Minor Group',
           'State',
           'Number of Individuals',
           'Taxable Income (mean)',
           'Taxable Income (median)',
           'Salary or wage income (mean)',
           'Salary or wage income (median)',
           'Total Income (mean)',
           'Total Income (median)',
              ]]
```
